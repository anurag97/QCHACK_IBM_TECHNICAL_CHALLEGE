```python
from qiskit.circuit.library.standard_gates import HGate,C3XGate
```

```python
qr = QuantumRegister(3)
qc1 = QuantumCircuit(qr)
c3h_gate = HGate().control(2)
qc1.append(c3h_gate, qr)
qc1.draw(output='mpl')
```

```python
qc_reduced.append(qc1,[0,1,2])
```

```python
qc_reduced.draw(output='mpl')
```

```python
qr2 = QuantumRegister(4)
qc2 = QuantumCircuit(qr2)
c3x_gate = C3XGate([0,1,2])
qc2.append(c3x_gate, qr2)
qc2.draw(output='mpl')
```

```python
qc_reduced.append(qc2,[0,1,2,3])
```

```python
qc_reduced.draw(output='mpl')
```
